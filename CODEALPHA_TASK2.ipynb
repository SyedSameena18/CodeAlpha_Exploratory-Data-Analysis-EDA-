## ***EXPLORATORY DATA ANALYSIS OF GLOBAL TECH LAYOFFS***

IMPORTNG LIBRARIES -

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING DATASET -

```python
import zipfile
with zipfile.ZipFile('/content/Layoffs.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')
```

```python
df = pd.read_csv('/content/Layoffs.zip')
```

CONVERTING DATE COLUMNS -



```python
df['date'] = pd.to_datetime(df['date'])
df['date_added'] = pd.to_datetime(df['date_added'])
df[['date','date_added']].head()
```

The `date` and `date_added` columns were converted from object data type to datetime format to enable time-based analysis and trend identification.

```python
df.info()
```



* The dataset contains 4,453 records and 11 columns.

  percentage_laid_off

*  Categorical columns:

   company, location, industry, stage, country, etc.




STATISTICAL SUMMARY -

```python
df.describe(include='all')
```



- The dataset contains information from 2,925 unique companies.
- The United States contributes the highest number of layoff records.
- Finance is the most frequently represented industry.
- Most layoff events were reported from the SF Bay Area.
- Post-IPO companies appear most frequently in the dataset.

```python
df.groupby('company')['total_laid_off'].sum().sort_values(ascending=False).head(10)
```

EXPLORATORY ANALYSIS -

```python
#Analysis 1: Top 10 Companies by Layoffs
top_companies = (
    df.groupby('company')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
top_companies
```

```python
#Analysis 2: Top 10 Countries by Layoffs
top_countries = (
    df.groupby('country')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
top_countries
```

```python
#Analysis 3: Top Industries by Layoffs
top_industries = (
    df.groupby('industry')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
top_industries
```

```python
#Analysis 4: Layoffs Trend by Year
df['year'] = df['date'].dt.year
df.groupby('year')['total_laid_off'].sum()
```

```python
#Analysis 5: Funding vs Layoffs
df[['total_laid_off', 'funds_raised']].corr()
```

```python
#Analysis 6: Companies with 100% Layoffs
df[df['percentage_laid_off'] == 1][
    ['company', 'industry', 'country']
].head(20)
```


1. The dataset contains 4,453 layoff records from 2,925 companies.
2. The United States accounts for the highest number of layoff events.
3. Certain industries experienced significantly higher layoffs than others.
4. Layoffs increased sharply during specific periods.
5. Several companies laid off 100% of their workforce.
